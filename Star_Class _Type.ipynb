## Data - Preprocessing

```python
# categorical variable convert into numeric form using pandas get_dummies()
df['Star color'].unique()
```

```python
color = pd.get_dummies(df['Star color'])
new_df = pd.get_dummies(df, columns = ['Star color'])
new_df
```

## Devide into target and independent variables

```python
red_val= color.columns[0]
red_val
```

```python
X = new_df.drop(columns=['Spectral Class'])
y= new_df['Spectral Class']
X,y
```

## train-test-split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y ,test_size = 0.3 , random_state = 42)
X_train
```

## Standardization of Data X_train,X_test

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# fit and transform train_x data
X_train_scaled = scaler.fit_transform(X_train)

# just taransform the test_x data ( secure from data-leakage)
X_test_scaled = scaler.transform(X_test)

X_train_scaled
```

## Apply different algorithms for test which algo apply well predict on dependent variable

### DecisionTree Algorithm

```python
# import DecisionTree Classifier

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()


# fit the model
model1 = dt.fit(X_train_scaled , y_train)

# predict the model
pred = dt.predict(X_test_scaled)
pred
```

### Accuracy score

```python
from sklearn.metrics import accuracy_score
accuracy =accuracy_score(y_test , pred)
print(f'Accuracy_Score : {accuracy}')
```

### RandomForest Classifier

```python
# import RandomForest Classifier

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()


# fit the model
model2 = rfc.fit(X_train_scaled , y_train)

# predict the model
pred = model2.predict(X_test_scaled)
pred
```

### Accuracy score

```python
from sklearn.metrics import accuracy_score
accuracy =accuracy_score(y_test , pred)
print(f'Accuracy_Score : {accuracy}')
```

# Logistic Regression

```python
# import Logistic Regression

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()


# fit the model
model3 = lr.fit(X_train_scaled , y_train)

# predict the model
pred = model3.predict(X_test_scaled)
pred
```

### Accuracy Score

```python
from sklearn.metrics import accuracy_score
accuracy =accuracy_score(y_test , pred)
print(f'Accuracy_Score : {accuracy}')
```

### K-Nearest Neighbor Algorithm

```python
# import K-Nearest-Neighbors

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()


# fit the model
model4 = knn.fit(X_train_scaled , y_train)

# predict the model
pred = model4.predict(X_test_scaled)
pred
```

### Accuracy_score

```python
from sklearn.metrics import accuracy_score
accuracy =accuracy_score(y_test , pred)
print(f'Accuracy_Score : {accuracy}')
```
